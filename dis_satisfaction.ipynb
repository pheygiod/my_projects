Let's now create a function to categorise the employees into four groups:
- New: Less than 3 years at the company
- Experienced: 3 to 6 years at the company
- Established: 7 to 10 years at the company
- Veteran: 11 or more years at the company

```python
# Converting the type of the series into float 
combined['institute_service'] = combined['institute_service'].astype(float)

# Creating the function
def service_map(val):
    if pd.isnull(val):
        return np.nan
    elif val < 3:
        return 'New'
    elif 3 <= val < 7:
        return 'Experienced'
    elif 7 <= val < 11:
        return 'Established'
    else:
        return 'Veteran'
    
# Applying the function to the series
combined['service_cat'] = combined['institute_service'].apply(service_map)

# Checking the data
combined['service_cat'].value_counts(dropna=False)
```
